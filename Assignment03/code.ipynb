```python
class Node:
    def __init__(self, player, parent=None, child=None):
        self.player = player
        self.parent = parent
        self.child = child if child is not None else []
        self.val = None

def tree(depth, player, parent=None):
    node = Node(player, parent)
    if depth == 0:        
        if player == 0:
            node.val = -1
        else:
            node.val = 1
    else:
        if player == 0:
            node.child = [Node(1, node), Node(1, node)]
        else:
            node.child = [Node(0, node), Node(0, node)]
        for i in node.child:
            tree(depth - 1, i.player, node)
    return node

def alpha_beta(node, alpha, beta, max_player):
    if node.val is not None:
        return node.val
    if max_player:
        value = -float('inf')
        for i in node.child:
            value = max(value, alpha_beta(i, alpha, beta, False))
            alpha = max(alpha, value)
            if alpha >= beta:
                break
        return value
    else:
        value = float('inf')
        for i in node.child:
            value = min(value, alpha_beta(i, alpha, beta, True))
            beta = min(beta, value)
            if alpha >= beta:
                break
        return value

def game(start):
    if start not in [0, 1]:
        print("Invalid")
        return
    game_tree = tree(5, start)
    round = 3
    win_list = []
    current_node = game_tree
    for _ in range(round):
        winner = alpha_beta(current_node, -float('inf'), float('inf'), current_node.player)
        if current_node.player == 0:
            if winner == -1:
                win_list.append("Scorpion")
            else:
                win_list.append("Sub-Zero")
        else:
            if winner == 1:
                win_list.append("Sub-Zero")
            else:
                win_list.append("Scorpion")
                      
        if current_node.child:
            current_node = current_node.child[0] if winner == -1 else current_node.child[1]
        else:
            pass

    print("Game Winner:", "Scorpion" if win_list[-1] == "Scorpion" else "Sub-Zero")
    print("Total Rounds Played:", len(win_list))
    for i in range(len(win_list)):
        print(f"Winner of Round {i+1}: {win_list[i]}")

def MortalCombat():
    start = int(input())
    game(start)

MortalCombat()



```

```python
# Task 02

def pacman_game(c):
    tree = {
        'root': {'player': 'Pac-Man', 'child': ['left', 'right']},
        'left': {'player': 'Ghost', 'child': ['left_left', 'left_right']},
        'right': {'player': 'Ghost', 'child': ['right_left', 'right_right']},
        'left_left': {'player': 'Pac-Man', 'value': 3},
        'left_right': {'player': 'Pac-Man', 'value': 6},
        'right_left': {'player': 'Pac-Man', 'value': 2},
        'right_right': {'player': 'Pac-Man', 'value': 7}
    }

    def alpha_beta(node, alpha, beta, player):
        if 'value' in node:
            return node['value']
        if player == 'Pac-Man':
            value = -float('inf')
            for child in node['child']:
                value = max(value, alpha_beta(tree[child], alpha, beta, 'Ghost'))
                alpha = max(alpha, value)
                if beta <= alpha:
                    break
            return value
        else:
            value = float('inf')
            for child in node['child']:
                value = min(value, alpha_beta(tree[child], alpha, beta, 'Pac-Man'))
                beta = min(beta, value)
                if beta <= alpha:
                    break
            return value

    
    root_val = alpha_beta(tree['root'], -float('inf'), float('inf'), 'Pac-Man')

    
    left_dm = 6 - c
    right_dm = 7 - c

    
    if left_dm > root_val or right_dm > root_val:
        if left_dm > right_dm:
            print("The new minimax value is", left_dm, ". Pacman goes left and uses dark magic.")
        else:
            print("The new minimax value is", right_dm, ". Pacman goes right and uses dark magic.")
    else:
        print("The minimax value is", root_val, ". Pacman does not use dark magic.")

pacman_game(2)
pacman_game(5)
```
